```python
import matplotlib.pyplot as grafico

cont =0;
listaMes = []
listaTempMedia = []
listaTempMinima = []
listaTempMaxima = []
with open("ex3 ia.txt", "r")as arquivo:
  texto = arquivo.read()
  textoSeparado = texto.split()
print(texto)
print("\n")
# print(textoSeparado)
# print(textoSeparado[30])
for textoSeparado in textoSeparado:

  if cont > 0 and cont < 13:
    # print(textoSeparado)
    listaMes.append(textoSeparado)

  if cont > 13 and cont < 26:
    # print(textoSeparado)
    listaTempMedia.append(textoSeparado)

  if cont > 26 and cont < 39:
    # print(textoSeparado)
    listaTempMinima.append(textoSeparado)

  if cont > 39:
    # print(textoSeparado)
    listaTempMaxima.append(textoSeparado)

  cont += 1

# print(f"Mes: {listaMes}")
# print(f"TempMedia: {listaTempMedia}")
# print(f"TempMinima: {listaTempMinima}")
# print(f"TempMaxima: {listaTempMaxima}")

###################################################################
########### Converção dos valores string para float ###############
###################################################################

listaTempMediaFloat = []
listaTempMediaFloat = list(map(float,listaTempMedia))
# print(listaTempMediaFloat)
listaTempMinimaFloat = []
listaTempMinimaFloat = list(map(float,listaTempMinima))
# print(listaTempMinimaFloat)
listaTempMaximaFloat = []
listaTempMaximaFloat = list(map(float,listaTempMaxima))
# print(listaTempMaximaFloat)
# print(cont)

###################################################################
############## Criação e implementação do dicionario ##############
###################################################################
dicMedia = {}
dicMinima = {}
dicMaxima = {}
for i in range(12):
  dicMedia[listaMes[i]] = listaTempMediaFloat[i]
  dicMinima[listaMes[i]] = listaTempMinimaFloat[i]
  dicMaxima[listaMes[i]] = listaTempMaximaFloat[i]

print("Temperatura Media do mes:")
print(dicMedia)
print("Temperatura Minima do mes:")
print(dicMinima)
print("temperatura Maxima do mes:")
print(dicMaxima)
###################################################################

###################################################################
################ Temperatura maxima do mes de julho ###############
###################################################################
print("")
print(f"Temperatura Media do {listaMes[6]} é {dicMedia['julho']}")
print("")
###################################################################
#################Criação e impreção dos graficos###################
###################################################################
grafico.figure(figsize=(12,8))
grafico.ylabel("Temperatura media/ºC")
grafico.xlabel("Mes")
grafico.grid()
grafico.title("Temperatura media mensal")
grafico.plot(listaMes , listaTempMediaFloat,marker='o')
##################################################################
grafico.figure(figsize=(12,8))
grafico.ylabel("Temperatura minima/ºC")
grafico.xlabel("Mes")
grafico.grid()
grafico.title("Temperatura minima mensal")
grafico.plot(listaMes , listaTempMinimaFloat,marker='o')
##################################################################
grafico.figure(figsize=(12,8))
grafico.ylabel("Temperatura maxima/ºC")
grafico.xlabel("Mes")
grafico.grid()
grafico.title("Temperatura maxima mensal")
grafico.plot(listaMes , listaTempMaximaFloat,marker='o')
```
